```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot log_2
x = np.arange(1, 1000, 1)
y_e = np.log2(x)
ax = plt.subplot()
sns.lineplot(y_e, label='log_2')
ax.legend()
plt.show()
```

```python
def TF(term_freq, max_count=200):
    # Term frequency of a term in a document
    # max_count = maximum occurences of any word in a document
    return term_freq / max_count

tf = np.arange(1, 200, 5)
plt.title('TF')
plt.plot(tf, TF(tf), '--')
plt.show()
```

```python
def IDF(doc_freq, num_docs=1000):
    # doc_freq = number of documents in which the term occurs (TF > 0)
    # Inverse Document Frequency
    # 0 if term is in every document
    # large if term is in few documents
    return np.log2(num_docs / doc_freq)

df = np.arange(1, 1000, 20)
plt.subplot(1,2,1)
plt.title('IDF')
plt.plot(df, IDF(df))
plt.subplot(1,2,2)
plt.title('IDF ohne log')
plt.plot(df, 1000/df)
plt.show()
```

```python
def TF_IDF(term_freq, doc_freq, num_docs=1000):
    return TF(term_freq) * IDF(doc_freq, num_docs)
```

```python
fig = plt.figure(figsize=(14, 5))
tf_range = np.arange(1, 2000, 1)
doc_freq_range = np.arange(1, 25, 1)
ax = plt.subplot(1,2,1)
plt.title('TF.IDF fr 1000 Dokumente')
tf = [0.05, 0.1, 0.2, 0.5, 0.75]
for t in tf:
    curr_tf_idf = []
    for df in doc_freq_range:
        curr_tf_idf.append(TF_IDF(t, df))

    plt.plot(doc_freq_range, curr_tf_idf)

plt.ylabel('TF.IDF')
plt.xlabel('Anzahl der Dokumente in der das Wort vorkommt')
plt.legend(tf , title='Vorkommen pro Dokument')
tf_range = np.arange(1, 2000, 1)


doc_freq_range = np.arange(1, 200, 1)
ax2 = plt.subplot(1,2,2)
plt.title('TF.IDF fr 1000 Dokumente')
tf = [0.05, 0.1, 0.2, 0.5, 0.75]
for t in tf:
    curr_tf_idf = []
    for df in doc_freq_range:
        curr_tf_idf.append(TF_IDF(t, df))

    plt.plot(doc_freq_range, curr_tf_idf)

plt.ylabel('TF.IDF')
plt.xlabel('Anzahl der Dokumente in der das Wort vorkommt')
plt.legend(tf , title='Vorkommen pro Dokument')
plt.show()

# for 10000 documents with 5000 words
fig = plt.figure(figsize=(14, 5))
doc_freq_range = np.arange(1, 25, 1)
ax = plt.subplot(1,2,1)
plt.title('TF.IDF fr 10000 Dokumente')
tf = [0.05, 0.1, 0.2, 0.5, 0.75]
for t in tf:
    curr_tf_idf = []
    for df in doc_freq_range:
        curr_tf_idf.append(TF_IDF(t, df, num_docs=5000))

    plt.plot(doc_freq_range, curr_tf_idf)

plt.ylabel('TF.IDF')
plt.xlabel('Anzahl der Dokumente in der das Wort vorkommt')
plt.legend(tf , title='Vorkommen pro Dokument')


doc_freq_range = np.arange(1, 200, 1)
ax2 = plt.subplot(1,2,2)
plt.title('TF.IDF fr 10000 Dokumente')
tf = [0.05, 0.1, 0.2, 0.5, 0.75]
for t in tf:
    curr_tf_idf = []
    for df in doc_freq_range:
        curr_tf_idf.append(TF_IDF(t, df, num_docs=5000))

    plt.plot(doc_freq_range, curr_tf_idf)

plt.ylabel('TF.IDF')
plt.xlabel('Anzahl der Dokumente in der das Wort vorkommt')
plt.legend(tf , title='Vorkommen pro Dokument')
plt.show()
```
